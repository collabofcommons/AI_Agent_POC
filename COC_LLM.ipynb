```python
import os
from langchain_experimental.agents import create_pandas_dataframe_agent
import pandas as pd
from langchain_huggingface import HuggingFaceEndpoint

# Path to your Kaggle dataset (update with actual path)
dataset_path = "./kaggle_dataset_folder/ipl_2025_auction_players.csv"

# Load the dataset into a Pandas DataFrame
try:
    df = pd.read_csv(dataset_path)
    print("Dataset loaded successfully!")
except FileNotFoundError:
    print(f"File not found: {dataset_path}")
    exit()

# Initialize the HuggingFaceEndpoint with the correct API token and model configuration
llm = HuggingFaceEndpoint(
    endpoint_url="https://api-inference.huggingface.co/models/meta-llama/Llama-2-7b-chat-hf",  # Replace with your model endpoint
    huggingfacehub_api_token="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",  # Replace with your API token
    temperature=0.7,
    model_kwargs={}
)

# Create a LangChain agent for the DataFrame
try:
    agent = create_pandas_dataframe_agent(llm, df, verbose=True, allow_dangerous_code=True)
    print("Agent created successfully!")
except Exception as e:
    print(f"Error creating agent: {e}")
    exit()

# Define your query
query = "Which 'Team' does player 'Virat Kohli' belong to? Respond only with the team name."

# Run the query on the agent
try:
    response = agent.invoke(query,  handle_parsing_errors=True)
    print("Response:", response)
except Exception as e:
    print(f"Error running query: {e}")
```
