```python
from langchain.chat_models import ChatOpenAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain

# Define your prompt template (or use a simple one if needed)
prompt = PromptTemplate(input_variables=["query"], template="Provide information on: {query}")

# Initialize the ChatOpenAI model
chatopenai = ChatOpenAI(model_name="o1-mini")
# chatopenai = ChatOpenAI(model_name="text-davinci-002")

# Initialize LLMChain with the model and prompt
llmchain_chat = LLMChain(llm=chatopenai, prompt=prompt)

# Run the chain with your input
result = llmchain_chat.run("podcast player")
print(result)
```
